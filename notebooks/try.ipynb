```python
import matplotlib.pyplot as plt

# Example input and attributions in [-1, 1]
tokens = ["What", "might", "Brad", "drink", "with", "eggs", "and", "toast", "at", "breakfast?"]
attributions = [-0.2, -0.6, 0.2, 0.3, -0.1, 0.8, 0.6, 0, -0.4, 0.25]

# Use orange for positive, blue for negative
colors = ['#1f77b4' if val > 0 else '#ff7f0e' for val in attributions]

plt.figure(figsize=(8, 2.5))
bars = plt.bar(tokens, attributions, color=colors)
plt.xticks(fontsize=14)
# plt.xticks(rotation=20, ha='right', fontsize=10)
# plt.ylabel("Attribution S0core", fontsize=11)
# plt.title("Attribution Over Input Tokens", fontsize=12)
# plt.ylim([-1, 1])
plt.axhline(0, color='gray', linewidth=0.8)
plt.yticks([])  # This hides the x-axis labels
plt.tight_layout()
```
